Custom styling for headers, automatic numbering of tables etc is added using a stylesheet.

It provides the formatting for headings, numbered figure captions and tables. 

#### Enable for <u>all</u> Notebooks

First, copy the ricardo_stylesheet.css file from the main GitHub repository. It is located here:
    
    https://github.com/genagibson/ricardo_template_style/
        styles/
            ricardo_stylesheet.css
            
On your computer, navigate to:

`~/.jupyter`

This is where the notebook files are stored. If there is already a folder called custom, save the ricardo_stylesheet.css in there. If not, create the folder. <b> Rename the css file to "custom.css"</b>. Your file structure should look like this:

    ~/.jupyter/
        custom/
            custom.css  # Important: rename the file to custom.css
            
Close down the notebook entirely, including the terminal, and restart to see the changes.

#### Enable it for a specific notebook

If for some reason, you don't want to set up the stylesheet for all Notebooks, or if you ever want to use a different css file, you can enable them for specific Notebooks.

To do this, copy the whole "styles" folder from this GitHub repository to your project folder: https://github.com/genagibson/ricardo_template_style/

### Interactive plotting

#### Libraries

We recommend using Plotly and Cufflinks for interactive plotting. 
* Plotly is an interactive plotting library
* Cufflinks binds Plotly to pandas dataframes

To get started, install the required packages

    pip install plotly
    pip install cufflinks 

When using these libraries to plot offline in Jupyter Notebooks, there is an additional initialisation step needed. Run the following to set up the notebook:

``` python
import plotly.plotly as py
import plotly.graph_objs as go
from plotly.offline import iplot, init_notebook_mode, offline
init_notebook_mode(connected=True)

import cufflinks as cf
cf.set_config_file(offline=True, 
                   world_readable=False, 
                   theme='white')
```

The extra commands are to enable offline plotting within the notebook.

#### Cufflinks for simple plots

Create the graph below using the following commands:

```python
display(HTML('<figcaption> Cufflinks graph - good for quick explorations </figcaption>'))
iris.iplot(x = 'sepal_length', y = 'sepal_width', 
           kind = 'scatter', mode = 'markers', size = 8,
          color = 'blue')
```

Cufflinks allows for creation of most plot types within a single line. You can also see that the automatic caption numbering can be used with any plotting library

````python
display(HTML('<figcaption> Cufflinks graph - good for quick explorations </figcaption>'))
iris.iplot(x = 'sepal_length', 
           y = 'sepal_width', 
           kind = 'scatter', 
           mode = 'markers', 
           size = 8,
          color = 'blue')
````

Cufflinks works well for most simple graphs and we recommend it for quick exploration. However, if you want to do more advanced customisations, you essentially need to access the underlying Plotly code - and depending on the amount of customisation you need, it might make sense to use Plotly directly. 

You can view the available parameters by typing: `help(iris.iplot)`



#### Plotly for advanced interactive plots

For more advanced customisation, you may need to use Plotly itself. The interface is quite simple:

```python
x = iris['sepal_length']
y = iris['sepal_width']
display(HTML('<figcaption> Plotly graph - better for customised finished graphs </figcaption>'))
fig = go.Figure()
fig.add_scatter(x=x,
                y=y,
                mode='markers',
                marker={'size': 10,
                        'opacity': 0.6,
                        'colorscale': 'Viridis'
                       });
iplot(fig)
```

````python
x = iris['sepal_length']
y = iris['sepal_width']

display(HTML('<figcaption> Plotly graph - better for customised finished graphs </figcaption>'))

fig = go.Figure()
fig.add_scatter(x=x,
                y=y,
                mode='markers',
                marker={'size': 10,
                        'opacity': 0.6,
                        'colorscale': 'Viridis'
                       });
iplot(fig)
````

Plotly graphs can be customised endlessly. Key functions to be aware of are:

### PDF

The built-in functions to save to PDF require pandoc and MikTeX installations, and tend to produce formatting errors. 

The best way to get a PDF is:

1. Save to HTML using either of the following code
```python
# To keep code visibility
jupyter nbconvert --to html notebook.ipynb
# To hide all code
jupyter nbconvert --template=nbextensions --to=html notebook.ipynb
```
2. Open the HTML file in your browser
3. Ctrl + P to print
4. Choose save as PDF
5. Choose More settings > untick headers and footers

This will "print what you see", so ensure the HTML file looks just as you want it. You will not see the numbered headings, as these come from the TOC2 extension and this won't be exported.

### Data protection and security risks

Clients may ask if html reports are "secure". This is something you should discuss with them right at the start of the project. 


````python

display(HTML('''<div class = "todo"> 
    <b> The data is stored in the HTML file itself and will not be secure. </b>
    The data will be relatively easy to access and change. <br>
    If the client wants to publish the HTML file but the data is sensitive, this is an obvious no-go <br>
    If the client wishes to have a confidential HTML report, we need to be mindful of how we share the files and use secure systems.
</div>'''))
````

### Hiding individual cells

Using the hide code button is the best way to remove all of the code in a notebook. You can also hide individual cells using the ^ button. Remember that you have to save the notebook before converting it to html in order to preserve the view.

However, the individually hidden code cells will all revert to being visible next time you open the notebook. So, if you want to mix hidden and visible code in your deliverable, the best way is to edit the cell metadata. This preserves the hidden and showing cells between sessions.

To do this:
1. go to View > Cell Toolbar > Edit Metadata
2. click the Edit Metadata button now showing to the top right of the cell
3. add 'hide_input': true to the JSON 
4. click edit to save
5. convert to html using `jupyter nbconvert 01_Ricardo_template_example.ipynb --template toc2`

You should pick one method or the other, but don't try to mix them as the hide commands from the extension and nbconvert can cancel out.
