# <center> Ricardo styles and custom table of contents
    
<b> <center> Setup instructions 
</b>

This notebook provides instructions to set up custom styles for Ricardo reporting in Jupyter, and a tour of the features.

## Step 1: Set up the stylesheet

Custom styling for headers, automatic numbering of tables etc is added using a stylesheet.

This is a css file saved in this repo:
    
    ricardo_template_style/
        files/
            styles/
                styles.css

You can copy this file to any project you are working on and activate it using the following commands:
  
    custom_style_file = 'files/styles/styles.css' # or the path to the style sheet
    styles = open(custom_style_file, 'r').read()
    HTML(styles)


### Numbering tables

Use style.set_caption to add numbered captions for tables. 

    table.style.set_caption('Example table')

The numbers will automatically increment.

```python
data = np.array([[1, 2, 3], [4, 5, 6]])
table = pd.DataFrame(data = data,
                    index = ('Row 1', 'Row 2'),
                    columns = ['Col 1', 'Col 2', 'Col 3'])

table.style.set_caption('Example table')
```

### Numbering figures

Since we might use various packages to create figures, the captions are created using HTML tag `<figcaption>`and automatically numbered in the same way as for tables.

To display HTML from within a code cell, we need to wrap the tag in display(HTML()) as follows:
    

    display(HTML('<figcaption> Example figure caption </figcaption>'))

### Internal comments

Several built-in styles are provided to make commenting easier. 

The first option is to put them in a code cell, which will render in HTML.


```python
# Code version
display(HTML('''<div class = "todo"> 
    Note: Things to do. This will be visible in exported HTML
</div>'''))
```

The second option is to include the tags in a markdown cell. This reduces typing but will not render in HTML, so it's only to be used for internal comments. 

Markdown version: 

<div class = 'commenter1'>
    Markdown: Comments from first author. 
    Markdown will not be rendered in HTML but this is fine for internal drafts
</div>


Other builtin options

<div class = 'commenter2'>
    Comments from second author
</div>

<div class = 'commenter3'>
    Comments from third author
</div>

## Step 4: Copy the custom folder 

### Copy the custom extensions folder across

The ricardo_custom_toc folder is saved in this repo:

### Reinstall the extensions

Jupyter looks in the folder for extensions and shows them on the extensions tab.

You need to reinstall nbextensions for the changes to take effect - run the following command in terminal.

    jupyter contrib nbextension install --user

<b> It will hang if you try running from the notebook. You should close everything and run it in the terminal instead. </b>

## Step 5: Enable the new extension
