## Step 2: Copy the custom folder 

### Copy the custom extensions folder across

The `ricardo_custom_toc` folder is saved in the same directory as this notebook.

Copy and paste the whole folder into the `jupyter_contrib_extensions` directory located in step 1. There you will see folders for all of the existing extensions. 

The file structure should now be:

    nbextensions/
        ricardo_custom_toc/
            -main.js
            -custom_toc.js
            -custom_toc.css
            -custom_toc.yaml
            


```python
Image('pictures/install_location.png', width = 600)
```

### Reinstall the extensions

Jupyter looks in the folder for extensions and shows them on the extensions tab.

You need to reinstall nbextensions for the changes to take effect - run the following command in terminal.

    jupyter contrib nbextension install --user

<b> It will hang if you try running from the notebook. You should close everything and run it in the terminal instead. </b>

## Step 3: Enable the new extension

Now that you've copied the files across, you should see `ricardo_custom_toc` in your extensions tab. If you don't see it, check that you reinstalled nbextensions.

Select Edit>nbextensions config and choose `ricardo_custom_toc`.

NB: <b> make sure you do not have Table of Contents (2) enabled </b>
