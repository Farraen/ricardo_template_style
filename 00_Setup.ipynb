

## Installing Python and Jupyter

The Jupyter Notebook App allows you to edit and run your notebooks in a web browser. The best way to get started is by installing Anaconda distribution, which gives you both Python and Jupyter Notebooks, plus most of the data analysis packages that you will need. 

* Follow the instruction guide here: https://docs.anaconda.com/anaconda/install/windows/ 
* Make sure you install Anaconda, not miniconda

You'll find everything under Anaconda Navigator in your start bar - it's worth pinning Jupyter Notebook to the taskbar at this point so that you can easily find it.
